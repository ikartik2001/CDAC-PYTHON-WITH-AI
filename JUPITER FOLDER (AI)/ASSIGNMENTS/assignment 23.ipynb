```python
import sklearn 
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
iris_df=pd.DataFrame(iris['data'],columns=iris["feature_names"])
```

```python
iris_df.head()
```

```python
# plotting graphs
```

```python
sns.pairplot(data=iris_df);
```

```python
# setting target and dependent variable
```

```python
x=iris['data']
y=iris['target']
```

```python
# data preprocessing
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=36)
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# model selection
```

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense
```

```python
model=keras.Sequential()
```

```python
model.add(Dense(50,activation='relu',input_shape=(x_train.shape[1],)))
model.add(Dense(40,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
from tensorflow.keras.optimizers import RMSprop
```

```python
optimizers = RMSprop(0.09) #here , we have set our learning rate
```

```python
model.compile(loss='binary_crossentropy',optimizer= optimizers,metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train,validation_split=0.3,epochs=300,batch_size=5,verbose=0)
```

```python
# now , we evaluate our model
```

```python
model.evaluate(x_test,y_test)
```

```python
# plotting the graphs
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['train','validation'])
plt.show()
```
