```python
#converting image to HSV format
hsv_image = cv2.cvtColor(file,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv_image)
```

```python
M = np.float32([[1,0,-100],[0,1,-100]]) 
dst = cv2.warpAffine(hsv_image,M,(100,200)) 
plt.imshow(dst)
```
