### create mupliple plot in one figrue

```python
# plt.subplot(nrows,ncols,plot_number)
plt.subplot(1,2,1)  #(row,col,plot_num)     
plt.plot(x,y,'r--')  # red with -- line pattern
plt.subplot(1,2,2)
plt.plot(y,x,'g*--');  # green with *-- pattern
```

## object oriented method inside matplotlib
### the object oriented methoid is to crate figure objects and then call methods or attributes

```python
# create figure 
fig = plt.figure()

# add set of axes to figure
axes = fig.add_axes([0.5,0.1,0.5,0.8]) # left bottom width height (range 0 to 1)

# plot on that set of axes
axes.plot(x,y,"g")
axes.set_xlabel('set x label') # notice the use  of set_ to begin method
axes.set_ylabel('se y lebel')
axes.set_title('set title')
```

```python
# create blank canvas
fig =plt.figure()

axes1= fig.add_axes([0.2,0.4,0.8,0.8])
axes2= fig.add_axes([1,0.5,0.6,1])

axes1.plot(x,y,'b')
axes1.set_xlabel('x1')
axes1.set_ylabel('y1')
axes1.set_title("Graph 1")

axes2.plot(x,y,'r')
axes2.set_xlabel('x1')
axes2.set_ylabel('y1')
axes2.set_title("Graph 2")
```

```python
# create blank canvas
fig, axes =plt.subplots()
axes1.plot(x,y,'r')
axes1.set_xlabel('x1')
axes1.set_ylabel('y1')
axes1.set_title("my subplot")

```

### dpi

```python
fig,axes =plt.subplots(figsize=(2,2),dpi=100)
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

### save figure method

```python
fig.savefig("mygraph.png",dpi=1000) # saving in png , 
fig.savefig("mygraph.pdf",dpi=10000)
fig.savefig("mygraph.tif")
```

## legends

```python
# you can use the label="label text" keyword argument when plots or other objects
# are added to the figure , and then using the legend method without arguments to add
# the legend to the figure:
    
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**4,label="x**2")
ax.plot(x,x**3,label="x**3")

ax.legend(loc=0)    # chaning legend loc 1,2,3,4 changes the postion of location of legend here and there
```

## setting color,linewidhts, linetypes

### matplot gives you a lot of options for coustomisng line types and bla bla

```python
# matlab style line color and style 
fig,ax= plt.subplots()
ax.plot(x,x**2,'b-') # blue line with dots

ax.plot(x,x**3,"g*-")
```

```python
fig,ax=plt.subplots()
ax.plot(x, x+1,color="blue")
ax.plot(x, x+3,color="yellow",alpha=0.5)
ax.plot(x, x+2,color="red",alpha=0.1)        # alpha adds transparency in the color
```

```python
# line width ... we can add in plot
fig,ax=plt.subplots()
ax.plot(x, x+1,color="blue",linewidth=11)
ax.plot(x, x+3,color="yellow",linewidth=11)
ax.plot(x, x+2,color="red",alpha=0.9,linestyle='--',marker="^",markersize=20,markerfacecolor="blue",markeredgecolor="orange")    # line space , markers , alpha and etc
```
