## count plot

```python
sns.countplot(x='size',hue='day',data=df,palette='magma');
```

```python
sns.countplot(x='size',hue='time',data=df,palette='magma')
```

```python
sns.pointplot(x='day',y='tip',data=df);
```

```python
sns.pointplot(x='day',y='tip',data=df)
sns.pointplot(x='time',y='total_bill',hue='smoker',data=df,palette='accent')
```

#  join plot

```python
join plot draws a plot of two variables with bivariate

sns.set_style('dark')
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```
