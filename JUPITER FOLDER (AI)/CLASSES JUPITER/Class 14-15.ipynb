### Line plots

```python
# a line plot is used to find relationship between x and y axis
```

```python
x1= np.linspace(0,10,100)
plt.plot(x1,np.sin(x1),'-',color='orange')
plt.plot(x1,np.cos(x1),'--',color='b')

plt.xlabel('x label')
plt.ylabel('y label')
```
