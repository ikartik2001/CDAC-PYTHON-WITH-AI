### Handeling the missing value

```python
clean.drop("Notes",axis=1,inplace=True)   # since notes has null values , its better to drop it .
clean
```

#### NOW DROPING THE STATUS ..WITH NULL VALUES

```python
clean.drop("Status",axis=1,inplace=True)
clean.head()
```

```python

clean.info()  # BASE PAY IS A OBJECT TYPE BECAUSE OF NOT PROVIDED DATA
```

```python
clean.drop(clean[clean["BasePay"]=="Not Provided"].index, inplace=True)
clean["BasePay"] = clean["BasePay"].astype(float)
clean.head()
```

```python
clean.isna().sum()
```

```python
clean.info()
```

```python
# change the data type of "basepay"
convert=["BasePay","OvertimePay","OtherPay","Benefits"]
for col in convert:
    clean[col]=clean[col].astype("float64")
    
clean.head()
clean.info()  # the datatype is changed form float to object
```

### How many people have job chief in their job title?

```python
title=sal['JobTitle']
```

```python


def chief_string(title):
    if 'chief'in title.lower():
        return True
    else:
        return False
```

```python
sal['JobTitle'].apply(chief_string).sum()  # applying function on sal job title using .apply() the sum
```
